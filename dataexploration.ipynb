```python
import pandas as pd
```

```python
dataset = pd.read_csv("dataset 4.csv")
dataset
```

```python
dataset.columns
```

```python

dataset.drop(columns=['Unemployment rate','Inflation rate', 'GDP'])
```

```python
print(dataset.isna().sum())
```

```python
import matplotlib.pyplot as plt
dataset.describe()

```

```python
import sklearn
from sklearn.model_selection import train_test_split
features = dataset.drop(columns=["Target"])
gt=dataset["Target"]
featuresTrain,featuresTest, gtTrain, gtTest = train_test_split(features,gt,test_size=0.2)
```

```python
featuresTrain
```

```python
from sklearn import tree
dt = tree.DecisionTreeClassifier()
dt = dt.fit(featuresTrain,gtTrain)
```

```python
tree.plot_tree(dt)
```

```python
ypred = dt.predict(featuresTest)
```

```python
sklearn.metrics.accuracy_score(gtTest, ypred)
```

```python
gtTest.describe()
```

```python
from sklearn.model_selection import GridSearchCV
dt2 = tree.DecisionTreeClassifier()
parametros = {'criterion':("gini", "entropy", "log_loss"),
                           'max_depth' : (None,5,10,15),'min_samples_split':(2,4,6)}
clf = GridSearchCV(dt2,parametros)
clf.fit(featuresTrain,gtTrain)
```

```python
clf.cv_results_
```

```python
yTest = clf.best_estimator_.predict(featuresTest)
```

```python
sklearn.metrics.accuracy_score(gtTest, yTest)
```

```python
f1 = sklearn.metrics.f1_score(gtTest,yTest,average=None)
```

```python
class_names= dataset.Target.unique()
print(class_names)
for i, f1s in enumerate (f1):
    class_name = class_names[i]
    print(f"F1 score for class {class_name} : {f1s}")
```
